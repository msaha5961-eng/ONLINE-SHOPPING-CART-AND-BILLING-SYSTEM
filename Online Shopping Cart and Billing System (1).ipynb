```python
# Online Shopping Cart & Billing System

class Product:
    def __init__(self, product_id, name, price, stock):
        self.product_id = product_id
        self.name = name
        self.__price = price      
        self.stock = stock
    def get_price(self):
        return self.__price
    def reduce_stock(self, quantity):
        self.stock -= quantity

class ShoppingCart:
    def __init__(self):
        self.cart = []
    def add_product(self, product, quantity):
        total = product.get_price() * quantity
        self.cart.append({
            "name": product.name,
            "price": product.get_price(),
            "quantity": quantity,
            "total": total
        })
        product.reduce_stock(quantity)
        print(f"\n{product.name} added to cart successfully!")
    def view_cart(self):
        if len(self.cart) == 0:
            print("\nYour cart is empty.")
            return
        print("\n   YOUR CART ")
        print("{:<15}{:<10}{:<10}{:<10}".format(
            "Product", "Price", "Qty", "Total"))
        print("-" * 50)
        for item in self.cart:
            print("{:<15}₹{:<9}{:<10}₹{:<10}".format(
                item["name"],
                item["price"],
                item["quantity"],
                item["total"]
            ))
        print("-" * 50)
        print("Subtotal: ₹", self.calculate_total())
    def calculate_total(self):
        total = 0
        for item in self.cart:
            total += item["total"]
        return total
products = [
    Product(1, "Rice", 60, 10),
    Product(2, "Sugar", 50, 8),
    Product(3, "Milk", 35, 15),
    Product(4, "Bread", 40, 10),
    Product(5, "Coffee", 180, 6),
    Product(6, "Biscuits", 30, 20)
]
cart = ShoppingCart()
def display_products():
    print("\n   PRODUCT CATALOG  ")
    print("{:<5}{:<15}{:<10}{:<10}".format(
        "ID", "Product", "Price", "Stock"))
    print("-" * 45)
    for product in products:
        print("{:<5}{:<15}₹{:<9}{:<10}".format(
            product.product_id,
            product.name,
            product.get_price(),
            product.stock
        ))
    print("-" * 45)
while True:
    print("\n ONLINE SHOPPING CART ")
    print("1. View Products")
    print("2. Add Product to Cart")
    print("3. View Cart")
    print("4. Next (Generate Bill)")
    print("5. Exit")
    try:
        choice = int(input("\nEnter your choice: "))
        if choice == 1:
            display_products()
        elif choice == 2:
            display_products()
            product_id = int(input("Enter Product ID: "))
            quantity = int(input("Enter Quantity: "))
            found = False
            for product in products:
                if product.product_id == product_id:
                    found = True
                    if quantity <= 0:
                        print("Quantity should be greater than 0.")
                    elif quantity > product.stock:
                        print("Sorry! Stock not available.")
                    else:
                        cart.add_product(product, quantity)
                    break
            if not found:
                print("Invalid Product ID.")
        elif choice == 3:
            cart.view_cart()
        elif choice == 4:
            print("\nProceeding to Billing...")
            break
        elif choice == 5:
            print("\nThank you for shopping!")
            break
        else:
            print("Invalid Choice.")
    except ValueError:
        print("Please enter numbers only.")
class Billing:
    def __init__(self, customer_name):
        self.customer_name = customer_name
        self.gst_rate = 0.05
    def generate_bill(self, cart):
        if len(cart.cart) == 0:
            print("\nYour cart is empty!")
            return
        subtotal = cart.calculate_total()
        gst = subtotal * self.gst_rate
        grand_total = subtotal + gst
        print("\n ---- SHOPPING BILL ----")
        print("Customer :", self.customer_name)
        print("-" * 40)
        print("{:<15}{:<10}{:<10}".format(
            "Product", "Qty", "Amount"))
        print("-" * 40)
        for item in cart.cart:
            print("{:<15}{:<10}₹{:<10}".format(
                item["name"],
                item["quantity"],
                item["total"]
            ))
        print("-" * 40)
        print(f"Subtotal      : ₹{subtotal:.2f}")
        print(f"GST (5%)      : ₹{gst:.2f}")
        print(f"Grand Total   : ₹{grand_total:.2f}")
        print("=" * 40)
        self.save_bill(cart, subtotal, gst, grand_total)
    def save_bill(self, cart, subtotal, gst, grand_total):
        with open("bill.txt", "w") as file:
            file.write("========== SHOPPING BILL ==========\n")
            file.write("Customer : " + self.customer_name + "\n")
            file.write("-" * 40 + "\n")
            for item in cart.cart:
                file.write(
                    f"{item['name']} | Qty: {item['quantity']} | ₹{item['total']}\n"
                )
            file.write("-" * 40 + "\n")
            file.write(f"Subtotal : ₹{subtotal:.2f}\n")
            file.write(f"GST (5%) : ₹{gst:.2f}\n")
            file.write(f"Grand Total : ₹{grand_total:.2f}\n")
        print("\nBill saved successfully in bill.txt")
customer_name = input("Enter Customer Name: ")
billing = Billing(customer_name)
while True:
    print("\nONLINE SHOPPING CART")
    print("1. View Products")
    print("2. Add Product to Cart")
    print("3. View Cart")
    print("4. Generate Bill")
    print("5. Exit")
    try:
        choice = int(input("\nEnter your choice: "))
        if choice == 1:
            display_products()
        elif choice == 2:
            display_products()
            product_id = int(input("Enter Product ID: "))
            quantity = int(input("Enter Quantity: "))
            found = False
            for product in products:
                if product.product_id == product_id:
                    found = True
                    if quantity <= 0:
                        print("Quantity should be greater than 0.")

                    elif quantity > product.stock:
                        print("Sorry! Stock not available.")
                    else:
                        cart.add_product(product, quantity)
                    break
            if not found:
                print("Invalid Product ID.")
        elif choice == 3:
            cart.view_cart()
        elif choice == 4:
            billing.generate_bill(cart)
        elif choice == 5:
            print("\nThank you for shopping!")
            break
        else:
            print("Invalid Choice.")

    except ValueError:
        print("Please enter numbers only.")
```
